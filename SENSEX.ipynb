```python
import sqlalchemy
import pandas as pd
import yfinance as yf
print('done')
```

```python
wiki = 'https://en.wikipedia.org/wiki/BSE_SENSEX'
```

```python
tickerSensex = pd.read_html(wiki)[1].Symbol.to_list()
```

```python
tickerSensex
```

```python
def getdata(tickers):
    data=[]
    for ticker in tickers:
        data.append(yf.download(ticker).reset_index())
    return data
```

```python
India = getdata(tickerSensex)
```

```python
India
```

```python
len(India)
```

```python
def createengine(name):
    engine=sqlalchemy.create_engine('sqlite:///'+name)
    return engine
```

```python
indiaengine = createengine('India')
indiaengine
```

```python
def TOSQL(frames,symbols,engine):
    for frame,symbol in zip(frames,symbols):
        frame.to_sql(symbol,engine,index=False)
        
    print("success in importing")
```

```python
TOSQL(India,tickerSensex,indiaengine)
```
